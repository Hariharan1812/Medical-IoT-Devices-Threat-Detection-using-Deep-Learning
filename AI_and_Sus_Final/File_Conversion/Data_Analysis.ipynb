## **Importing Required Libraries**


```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Data Visualization**

*   Importing the datset from the local library.
*   Implementing several visualization techniques to visualize the Dataset and analysing the nature of the dataset.



```python
# Paths of dataset folders
train_path = "<Enter the train dataset file location>"
test_path = "<Enter the test dataset file location>"


# Load all parquet files in the train folder
train_df = pd.concat([
    pd.read_parquet(os.path.join(train_path, file))
    for file in os.listdir(train_path) if file.endswith('.parquet')
], ignore_index=True)

# Load all parquet files in the test folder
test_df = pd.concat([
    pd.read_parquet(os.path.join(test_path, file))
    for file in os.listdir(test_path) if file.endswith('.parquet')
], ignore_index=True)


```

### **Dataset Information**

```python
# Shows structure of the train data
print(train_df.info())
print(train_df.head())
```

### **Dataset Summary**

```python
# Get list of numeric columns
numeric_cols = train_df.select_dtypes(include='number').columns.tolist()

# Show summary statistics
print("\nSummary Statistics:")
print(train_df[numeric_cols].describe())
```

### **Dataset Pairwise Realtionship**

```python
# Get list of numeric columns
numeric_cols = train_df.select_dtypes(include='number').columns.tolist()

#Plot the Pairwise Relationship Graph
features_to_plot = numeric_cols[:6]
sns.pairplot(train_df[numeric_cols[:4]])
plt.suptitle("Pairwise Relationships", y=1.02)

# Display the plot
plt.show()
```
